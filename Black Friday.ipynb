## Step 1 :: Read csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Black_friday_train.csv")
```

## Step 2 :: High level analysis

```python
df.head()
# df.info()
# df.describe()
#df.describe(include='object')
```

## Step 3 :: Null Check

```python
null_percent = (df.isnull().sum() / (df.shape[0])*100).sort_values(ascending = False)
print(null_percent)
```

## Step 4 :: Null correction

```python
# mv = df['Product_Category_3'].mean()
# mv
df.fillna(0, inplace = True)
```

## Step 5 :: Label encoding and One hot encoding

```python
from sklearn.preprocessing import LabelEncoder
la = LabelEncoder()
df['Product_ID'] = la.fit_transform(df['Product_ID'])
# df.describe(include='object')

df['Gender'] = np.where(df['Gender'] == 'M' ,1,0)
# df['Gender'] = df['Gender'].replace(["M","F"], [1,0])

df = pd.get_dummies(df)
df
```

```python
df.info()
```

## Step 6 :: Feat Engg by corr heatmap, plot, describe, random select etc

```python
import seaborn as sns

corr = df.corr()
corr = np.abs(corr)
# plt.figure(figsize=(20,16))
# sns.heatmap(corr, annot=True)
# plt.show()

corr['Purchase'].sort_values()
# df.drop(['Marital_Status'], axis='columns')
```

## Step 7 :: Separating the dependent and independent variables

```python
x = df.drop(['Purchase'], axis='columns')
y = df['Purchase']
```

## Step 8 :: Train and Test split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 100)
```

```python
x_train.shape, x_test.shape
```

# Step 9 :: Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x_train = scaler.transform(x_train)
x_test  = scaler.transform(x_test)
```

## Step 10 :: Baseline Model

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()

model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
```

## Step 11 :: Hyperparameter tunning

```python
from sklearn.model_selection import RandomizedSearchCV
model = RandomForestRegressor(n_jobs=-1)

n_estimators = [int(x) for x in np.linspace(start = 1, stop = 200, num=4)]
max_depth    = [int(x) for x in np.linspace(start = 1, stop = 50, num=11)]
max_depth.append(None)
max_features = ['auto','sqrt']
min_samples_leaf =[1,2,4]
max_leaf_nodes =[2,3,5,10]
bootstrap=[True,False]

random_grid = { 'n_estimators': n_estimators,
                'max_depth'   : max_depth,
                'max_features': max_features,
                'min_samples_leaf': min_samples_leaf,
                'max_leaf_nodes' : max_leaf_nodes,
                'bootstrap'   : bootstrap
    
}

rf_random = RandomizedSearchCV(estimator = model, param_distributions = random_grid, cv = 3)
rf_random.fit(x_train,y_train)
```

```python
rf_random.best_params_
```

```python
model = rf_random.best_estimator_
model.fit(x_train,y_train)
print(model.score(x_test,y_test))
y_pred = model.predict(x_test)
print(y_test[:10])
print(y_pred[:10])
```
